### Q1:  Show the result of PG in terms of ||w(t) − wˆ|| w.r.t. the number of iteration

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[3.0, 0.5],
             [0.5, 1.0]])
mu = np.array([1, 2])
```

```python
def ST(q, mu):
    if mu>q:
        return mu-q
    elif mu<-q:
        return mu+q
    else:
        return 0
```

```python
eta = 1/np.max(np.linalg.eig(2*A)[0])
```

```python
lambs = np.array([2, 4, 6])
w = np.array([3, -1],dtype=float)
for lamb in lambs:
    for i in range(0,30):
        temp0 = w[0]-eta*(2*np.dot((w-mu), A)[0])
        temp1 = w[1]-eta*(2*np.dot((w-mu), A)[1])
        w[0] = ST(eta*lamb, temp0)
        w[1] = ST(eta*lamb, temp1)
    print('w = ({:.2f}, {:.2f}).T'.format(w[0],w[1]))
```

```python
iteration = []
for i in range(30):
    iteration.append(i)
```

```python
#λ=2
w = np.array([3, -1],dtype=float)
diff_2 = []
for i in range(0,30):
    temp0 = w[0]-eta*(2*np.dot((w-mu), A)[0])
    temp1 = w[1]-eta*(2*np.dot((w-mu), A)[1])
    w[0] = ST(eta*2, temp0)
    w[1] = ST(eta*2, temp1)
    diff_2.append(np.linalg.norm(w-[0.82, 1.09], ord = 2))
```

```python
#λ=4
w = np.array([3, -1],dtype=float)
diff_4 = []
for i in range(0,30):
    temp0 = w[0]-eta*(2*np.dot((w-mu), A)[0])
    temp1 = w[1]-eta*(2*np.dot((w-mu), A)[1])
    w[0] = ST(eta*4, temp0)
    w[1] = ST(eta*4, temp1)
    diff_4.append(np.linalg.norm(w-[0.64, 0.18], ord = 2))
```

```python
#λ=6
w = np.array([3, -1],dtype=float)
diff_6 = []
for i in range(0,30):
    temp0 = w[0]-eta*(2*np.dot((w-mu), A)[0])
    temp1 = w[1]-eta*(2*np.dot((w-mu), A)[1])
    w[0] = ST(eta*6, temp0)
    w[1] = ST(eta*6, temp1)
    diff_6.append(np.linalg.norm(w-[0.33, 0.00], ord = 2))
```

```python
plt.figure()
plt.yscale('log')
plt.plot(iteration, diff_2,label='lambda=2')
plt.plot(iteration, diff_4,label='lambda=4')
plt.plot(iteration, diff_6,label='lambda=6')
plt.xlabel('iteration')
plt.ylabel('||w(t) - w^||')
plt.legend()
```

### Q2:  Investigate concept of regularization path and plot regularization path of this lasso

```python
#Regularization path： λを∞から0まで変化させたときのβ（λ）の軌跡
```

```python
lambs = np.array([0.1, 0.2, 0.5, 1, 2, 4, 6, 8])
w_lamb=[]
    
for lamb in lambs:
    w = np.array([3, -1],dtype=float)
    for i in range(0,30):
        temp0 = w[0]-0.16*(2*np.dot((w-mu), A)[0])
        temp1 = w[1]-0.16*(2*np.dot((w-mu), A)[1])
        w[0] = ST(0.16*lamb, temp0)
        w[1] = ST(0.16*lamb, temp1)
    w_lamb.append([w[0],w[1]])
w_lamb = np.array(w_lamb)

plt.figure
plt.plot(w_lamb[:,0],w_lamb[:,1],marker='o')
x = w_lamb[:,0]
y = w_lamb[:,1]
txt = ['λ=0.1','λ=0.2','λ=0.5','λ=1','λ=2','λ=4','λ=6','λ=8']
plt.scatter(x, y)
plt.xlabel("w1")
plt.ylabel("w2")
for i in range(len(x)):
    plt.annotate(txt[i], xy = (x[i], y[i]), xytext = (x[i]+0.1, y[i]+0.1))
plt.show()

lambdas = ['0.1','0.2','0.5','1','2','4','6','8']
plt.figure()
plt.plot(lambdas, x, label='w1')
plt.plot(lambdas, y, label='w2')
plt.xlabel('lambda')
plt.ylabel('w')
plt.legend()
```
