# Word embeddings trained on Japanese Wikipedia articles

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import gensim

def plot_vector(x, opt='b'):
    plt.xlim(0, len(x))
    plt.plot(range(len(x)), x, opt)
```

## Loading word embeddings

```python
m = gensim.models.KeyedVectors.load_word2vec_format('/home/okazaki/install/jawiki.bin', binary=True, unicode_errors='ignore')
```

## Visualizing word vectors

```python
m['栃木県']
```

```python
len(m['栃木県'])
```

```python
plot_vector(m['栃木県'], 'b')
```

```python
plot_vector(m['群馬県'], 'b')
```

```python
plot_vector(m['コンピュータ'], 'b')
```

## Computing the similarity between two word vectors

```python
m.similarity('栃木県', '群馬県')
```

```python
m.similarity('栃木県', 'コンピュータ')
```

## Finding the most similar words

```python
m.most_similar(positive=['栃木県'])
```

```python
m.most_similar(positive=['コンピュータ'])
```

```python
m.most_similar(positive=['嬉しい'])
```

```python
m.most_similar(positive=['さんま'])
```

```python
m.most_similar(positive=['サンマ'])
```

## Solving the word analogy task

```python
m.most_similar(positive=['宇都宮', '群馬県'], negative=['栃木県'])
```

```python
m.most_similar(positive=['安倍晋三', 'ドイツ'], negative=['日本'])
```
