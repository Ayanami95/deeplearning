# Word embeddings trained on English News articles

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import gensim

def plot_vector(x, opt='b'):
    plt.xlim(0, len(x))
    plt.plot(range(len(x)), x, opt)
```

## Loading word embeddings

```python
m = gensim.models.KeyedVectors.load_word2vec_format('/home/okazaki/install/GoogleNews-vectors-negative300.bin', binary=True)
```

## Visualizing word vectors

```python
m['Manchester']
```

```python
len(m['Manchester'])
```

```python
plot_vector(m['Manchester'])
```

```python
plot_vector(m['Liverpool'])
```

```python
plot_vector(m['computer'])
```

## Computing the similarity between two word vectors

```python
m.similarity('Manchester', 'Liverpool')
```

```python
m.similarity('Manchester', 'computer')
```

## Finding the most similar words

```python
m.most_similar(positive=['Manchester'])
```

```python
m.most_similar(positive=['water'])
```

## Solving the word analogy task

```python
m.most_similar(positive=['woman', 'king'], negative=['man'])
```

```python
m.most_similar(positive=['Paris', 'UK'], negative=['France'])
```
